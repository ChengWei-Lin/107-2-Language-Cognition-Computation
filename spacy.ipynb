```python
import glob, re, os, nltk, collections, math, spacy, time
import pandas as pd
nlp = spacy.load('en_core_web_md')
```

```python
path = os.getcwd()
cand_li = os.listdir(path+"\data")
direc0 = []
for cand in cand_li:
    direc0.append(os.listdir(path+"\data\{}\\rawdata".format(cand)))
for a in range(len(direc0)):
    for b in range(len(direc0[a])):
        direc0[a][b] = path+"\data\\" + cand_li[a] + "\\rawdata\\" + direc0[a][b]
direc = []
for a in range(len(direc0)):
    for b in range(len(direc0[a])):
        direc.append(direc0[a][b])
```

```python
import pandas as pd
import spacy
import time
start = time.time()
nlp = spacy.load("en_core_web_sm")

df = pd.DataFrame()

for a in range(5):
    with open(direc[a], 'r', encoding='utf-8') as f:
        f = f.read()
        f = f.replace('\n', ' ')
        #f = f.replace('\'', ' ')
        #f = re.sub('\s(re)\s', ' are ', f)
        f = re.sub('\s+', ' ', f)
    doc = nlp(f)
    for token in doc:
        d = {#'_CHILDREN' : [child for child in token.children], 
             #'_HEAD POS' : token.head.pos_,
             '7_isStop' : token.is_stop,
             '6_isAlpha' : token.is_alpha,
             '5_SHAPE' : token.shape_,
             '4_DEP' : token.dep_,
             '3_TAG' : token.tag_,
             #'_HEAD TEXT' : token.head.text, 
             '2_POS' : token.pos_,
             '1_LEMMA' : token.lemma_,
             '0_TEXT' : token.text}
        df = df.append(d, ignore_index=True)
end = time.time()
print('Time: {} mins. '.format((end-start)/60))
```

```python
import spacy
nlp = spacy.load('en_core_web_md')
tokens = nlp(u'dog cat banana afskfsd')

for token in tokens:
    print(token.text, token.has_vector, token.vector_norm, token.is_oov)
```

```python
tokens = nlp(u'dog cat banana')

for token1 in tokens:
    for token2 in tokens:
        print(token1.text, token2.text, token1.similarity(token2))
```
